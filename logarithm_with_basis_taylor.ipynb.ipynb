Приближенное вычисление логарифма по заданному основания при помощи ряда Тейлора

```python
import math

def ln(n, eps):
    assert n > 0
  
    x = n - 1

    if x == 0:
        return 0

    elif abs(x) < 1:
        i = 2;
        element = x
        sum = x
        while abs(element) > eps:
            element = pow(-1, i - 1) * pow(x, i) / i
            sum += element
            i += 1
            
        return sum
    
    alpha = 0
    ln10 = 2.3025851 # для х>0
    
    while n >= 1:
        n /= 10
        alpha += 1
        
    return (ln(n, eps) + alpha * ln10)
```

Зададим основание и значение логарифма и вычислим его

```python
basis = input()
value = input()
print("log      = " + str(ln(int(value), 1e-7)/ln(int(basis), 1e-7)))
```

Сравним результат с результатом библиотечной функции Math.log

```python
print("Math.log = " + str(math.log(int(value))/math.log(int(basis))))  
```
